```python
# check if there are null values in the data
df.isnull().sum()
```

```python
# as we can see only the numerical columns has the missing values
# so we need the fill the missing cells using either mean or median
# But mean can cause error if there is outliers present in the data so first check if their is outlier or not
plt.figure(figsize=(15,5))

# 1st chart
plt.subplot(1,3,1)

sns.boxplot(
    x=df['Number of Guests']
)

plt.title("Guests")


# 2nd chart
plt.subplot(1,3,2)

sns.boxplot(
    x=df['Quantity of Food']
)

plt.title("Food Quantity")


# 3rd chart
plt.subplot(1,3,3)

sns.boxplot(
    x=df['Wastage Food Amount']
)

plt.title("Food Wastage")


plt.tight_layout()

plt.show()
```

```python
# As we can see their are outliers present therfore median is optimal way to fill the empty values
numeric_cols = ['Number of Guests',	'Quantity of Food',	'Wastage Food Amount']
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
```

```python
#As we can see in the above array the values haves different cases and spaces
# we need to convert all the values into small case and trim the spaces
df['Event Type']= (df['Event Type'].str.lower().str.strip())
```

```python
object_cols = [
    'Type of Food',
    'Event Type',
    'Storage Conditions',
    'Purchase History',
    'Seasonality',
    'Preparation Method',
    'Geographical Location',
    'Pricing'
]

for col in object_cols:
    print(col)
    print(df[col].unique())
    print()
```

```python
columns_to_clean = [
    'Type of Food',
    'Storage Conditions'
]

for col in columns_to_clean:
    
    df[col] = (
        df[col]
        .str.lower()
        .str.strip()
    )
```
